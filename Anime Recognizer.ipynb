# Anime Recognizer

This project is part of lesson 1 of the FastAI introduction to AI course. The assignment is to find a creative use for a vision or tabular deep learning model. I have decided to create an image classifier which can recognize a set of anime shows from screenshots of an episode of the show.

The goal of this project is creating a trained model that recognizes stylistic features from anime shows. This is more challenging than a simple object classification task because it must understand styles applied to new subjects as opposed to simply recognizing subjects.

## Training data
For our training data set, we will use square images taken from cropped screenshots of video files of different anime. We will write a python script that reads in video files (episodes of anime shows), seeks to 10 second increments of the video, and takes cropped screenshots from the center of the frame.

To keep training data balanced, we will create a set amout (set below to 600) screenshots for each anime.

## Model and training
We will try two approaches with the resnet34 architecture. First, we will fine-tune a resnet34 model pre-trained on IMAGENET1K_V2. Then we will train a blank resnet34 architecture and compare the error rate of each to determine the best approach.

````python
from pathlib import Path

def create_dir_if_not_exists(path_str: str):
    path = Path(path_str)
    path.mkdir(parents=True, exist_ok=True)

training_data_dir = "training_data/anime_videos/"
cropped_screenshots_output_dir = "training_data/anime_screenshots_cropped/"

# Create required directories if they don't exist
create_dir_if_not_exists(training_data_dir)
create_dir_if_not_exists(cropped_screenshots_output_dir)
````

````python
import os

# Determine which shows are available to train on
dir_entries = os.listdir(training_data_dir)
show_slugs = [entry for entry in dir_entries if os.path.isdir(os.path.join(training_data_dir, entry))]
print("Found slugs for the following shows:")
for slug in show_slugs:
    print(slug)
````

````python
import ffmpeg
from PIL import Image
from io import BytesIO
from pathlib import Path

# For each show, load the videos, take screenshots, then unload the video file

def crop_to_square(image_in_memory: BytesIO) -> BytesIO:
    with Image.open(image_in_memory) as img:
        width, height = img.size
        new_size = min(width, height)

        left = (width - new_size) / 2
        top = (height - new_size) / 2
        right = (width + new_size) / 2
        bottom = (height + new_size) / 2

        cropped_img = img.crop((left, top, right, bottom))
        cropped_img_io = BytesIO()
        cropped_img.save(cropped_img_io, format='PNG')
        cropped_img_io.seek(0)
        return cropped_img_io

def resize_image(image_in_memory: BytesIO) -> BytesIO:
    with Image.open(image_in_memory) as img:
        resized_img = img.resize((200, 200))
        resized_img_io = BytesIO()
        resized_img.save(resized_img_io, format='PNG')
        resized_img_io.seek(0)
        return resized_img_io

def save_screenshot(video_filepath: str, output_dir: str, output_filepath: str, second: int) -> None:
    # Extract the frame
    out, _ = (
        ffmpeg
        .input(video_filepath, ss=second)
        .output('pipe:', format='image2', vframes=1)
        .run(capture_stdout=True, capture_stderr=True)
    )
    image_in_memory = BytesIO(out)

    # Crop and resize the image
    cropped_image = crop_to_square(image_in_memory)
    resized_image = resize_image(cropped_image)

    # Create the output directory if it doesn't already exist
    path = Path(output_dir)
    path.mkdir(parents=True, exist_ok=True)

    # Save the final image to disk
    with open(output_filepath, 'wb') as f:
        f.write(resized_image.getbuffer())

def save_screenshots_from_video(video_filepath: str, slug: str, max_screenshots: int=100) -> int:
    """Creates and saves up to max_screenshots screenshots from the given video, returning the number created."""
    print(f"Attempting to create screenshots from video: {video_filepath}")

    # Use ffmpeg probe to get video duration
    try:
        probe = ffmpeg.probe(video_filepath)
        duration = float(probe['format']['duration'])
    except ffmpeg.Error as e:
        print("FFprobe Error:", e.stderr.decode())
        raise e

    # Save a screenshot every 10 seconds
    interval = 10
    output_dir = os.path.join(cropped_screenshots_output_dir, slug)
    video_filename = os.path.basename(video_filepath)
    for screenshot_index, second in enumerate(range(0, int(duration), interval)):
        if screenshot_index == max_screenshots:
            print(f"Maximum number of screenshots created. ({max_screenshots}) Skipping remaining.")
            return screenshot_index
        output_filepath = os.path.join(output_dir, f"{video_filename.replace('.', '_')}_screenshot_{second}sec.png")
        print(f"Saving screenshot for {os.path.basename(output_filepath)}... ", end="")
        # Check if the output file exists already, and skip it if so
        if os.path.isfile(output_filepath):
            print("File already exists. Skipping.")
            continue
        save_screenshot(video_filepath, output_dir, output_filepath, second)
        print("Done.")
    return max_screenshots
        

video_extensions = ['.mp4', '.avi', '.mov', '.wmv', '.flv', '.mkv']

def save_screenshots_for_slug(slug: str, max_screenshots=600, max_screenshots_per_video=100):
    videos_dir = os.path.join(training_data_dir, slug)
    print(f"Creating screenshots for videos in {videos_dir}")
    video_filenames = os.listdir(videos_dir)
    video_filenames = [filename for filename in video_filenames if os.path.splitext(filename)[1].lower() in video_extensions]
    video_filenames.sort()
    created_screenshots = 0
    for video_filename in video_filenames:
        if created_screenshots == max_screenshots:
            print(f"Created mamimum number of screenshots for slug {slug}. ({max_screenshots}). Skipping remaining videos.")
            break
        created_screenshots += save_screenshots_from_video(os.path.join(videos_dir, video_filename), slug, max_screenshots=min(max_screenshots-created_screenshots, max_screenshots_per_video))
    print(f"Done creating screenshots for videos in {videos_dir}")

for slug in show_slugs:
    save_screenshots_for_slug(slug)
    
print("Done creating screenshots for all videos.")
````

````python
from fastai.vision.all import *

dls = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label
).dataloaders("training_data/anime_screenshots_cropped/", bs=32)

dls.show_batch(max_n=6)
````
